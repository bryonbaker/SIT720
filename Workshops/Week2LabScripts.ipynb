```python
# note, we need a square matrix (# rows = # cols), use C:
import numpy as np

matOne = np.random.randn(3,3)
matTwo = np.random.randn(3,3)

print(matOne)

matOneInverse=np.linalg.inv(matOne)
print(matOneInverse)


divMat=matOneInverse.dot(matTwo)
print(divMat)


newMatOne=np.linalg.inv(matOneInverse)
print(newMatOne)

```
